```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_wallet_balance(df, wallet_id, start_date_str, n_days=7, shift_days=0):
    """
    Plot accumulated token balance for a wallet over a specific time window.

    Parameters
    ----------
    df : pd.DataFrame
        Must contain columns ['wallet_id', 'time', 'is_ask', 'size'].
    wallet_id : str
        ID of the wallet to plot.
    start_date_str : str
        Start date in string format, e.g. '2025-01-01'.
    n_days : int, default=7
        Number of days forward from start_date to include in the plot.
    shift_days : int, default=0
        Number of days to shift the window forward or backward.
        Positive values move the window forward, negative values backward.
    """

    # Ensure time column is datetime
    df['time'] = pd.to_datetime(df['Date'])

    # Apply window logic
    start_date = pd.to_datetime(start_date_str) + pd.Timedelta(days=shift_days)
    end_date = start_date + pd.Timedelta(days=n_days)

    # Filter for wallet and date range
    mask = (df['User_Id'] == wallet_id) & (df['time'].between(start_date, end_date))
    df_wallet = df.loc[mask].copy()

    if df_wallet.empty:
        print(f"No data found for wallet {wallet_id} between {start_date.date()} and {end_date.date()}.")
        return

    # Sort by time
    df_wallet = df_wallet.sort_values('time')

    # Compute signed trade size
    df_wallet['signed_size'] = np.where(df_wallet['Type'] == "buy", -df_wallet['Bitcoins'], df_wallet['Bitcoins'])

    # Compute accumulated balance
    df_wallet['balance'] = df_wallet['signed_size'].cumsum()

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(df_wallet['time'], df_wallet['balance'], marker='o', linewidth=2)
    plt.title(f"Accumulated Token Balance\nWallet: {wallet_id}\n{start_date.date()} → {end_date.date()}")
    plt.xlabel("Time")
    plt.ylabel("Accumulated Balance")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
plot_wallet_balance(df, wallet_id, start_date_str, n_days=7, shift_days=0)
```
